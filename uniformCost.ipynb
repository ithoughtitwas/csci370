This is the uniform cost search. It is somewhat more complex than a breadth-first or depth-first search in that the cost of paths need to be tracked.

```python
import copy

# Teo example search trees
#
#     S             S
#   3/ \2        1/   \4
#   A   B        A     B 
#  1|   |6     3/ \2   |5
#   G   G     C    D   G
#            5|  9/ \3 
#             E  F   G
#            5|
#             G

searchSpace1 = [('S', 'A', 3), ('S', 'B', 2), ('A', 'G', 1), ('B', 'G', 6)]

searchSpace2 = [('S', 'A', 1), ('A', 'C', 3), ('C', 'E', 5),
('E', 'G', 5), ('A', 'D', 2), ('D', 'F', 9), ('D', 'G', 3),
('S', 'B', 4), ('B', 'G', 5)]

# Used to sort a list of tuples on the second element.
def GetKey(item):
   return item[1]

# Collects into a list all children of a node.
def GetChildren(space, node):
   list = [item for item in space if item[0] == node]
   return list
```
