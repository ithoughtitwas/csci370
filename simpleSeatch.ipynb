The purpose of this notebook is to demonstrate a simple breadth-first search. Nothing fancy at all. The key is to inderstand how the queue is bring used to track the order that nodes are to be visited.

```python
# This is an example search tree
#
#     S
#   /   \
#  A     B
#  |   / | \
#  C  D  E  F
#  |  |     |
#  H  G     I

searchTree = [
   ('S', ['A', 'B']),
   ('A', ['C']),
   ('C', ['H']),
   ('B', ['D', 'E', 'F']),
   ('D', ['G']),
   ('F', ['I']),
   ('H', []),
   ('G', []),
   ('E', []),
   ('I', [])
]

def GetChildren(node):
   for k in range(len(searchTree)):
      if node == searchTree[k][0]:
         return searchTree[k][1]
```

```python
queue = ['S']
goalNode = 'G'
found = False
while not found and queue != []:
   print(queue)
   if queue[0] == goalNode:
      found = True
   else:
      children = GetChildren(queue[0])
      queue = queue[1:] + children

if found:
   print('Goal found!')
else:
   print('Goal not found')
```
