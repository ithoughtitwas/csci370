The purpose of this notebook is to show a simple sequential Keras model with a CSV dataset. Scaling is done via Scikit-Learn. 

```python
# Import needed libraries

from google.colab import files
import io
import keras
import numpy as no
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
```

```python
# Use Pandas to input a local dataset

# uploaded = files.upload()
# myPandas = pd.read_csv(io.BytesIO(uploaded['someData.csv']))

url = 'https://raw.githubusercontent.com/ithoughtitwas/csci370/main/data/carBuyData.csv'
myPandas = pd.read_csv(url)

print(myPandas)

# Convert the Pandas dataframe to a Numpy array

myNumpy = myPandas.to_numpy()
print(myNumpy)

# Separate features from targets

myX = myNumpy[:, 0:2] # attributes
myY = myNumpy[:, 2].flatten() # target

myScaler = MinMaxScaler()
scaleX = myScaler.fit_transform(myX)

print(scaleX)
print(myY)
```

```python
model = keras.models.Sequential()
model.add(keras.layers.Dense(50, input_shape=(2,), activation="relu"))
model.add(keras.layers.Dense(10, activation="relu"))
model.add(keras.layers.Dense(1, activation="sigmoid"))

print(model.summary())
```

```python
model.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
history = model.fit(scaleX, myY, epochs=10)
```
