This notebook is a first  intro to using the MNIST dataset. No ML here but just input the dataset and output some images. Also, FMNIST is introduced.

You will notice that the input of the two datasets is different. This is because MNIST is opened with Scikit-Learn and FMNIST is opened with Keras.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_openml

import keras
import tensorflow as tf
```

It can take a few seconds to input the dataset; thus, the input is in a different code cell to avoid inputting it more than needed.

```python
mnist = fetch_openml('mnist_784', version=1)
```

In the following code, try outputting different images (digits). Change the 4 to whatever digit you wish to examine (starting with index 0 for the first digit in the dataset).

```python
X, y = mnist["data"], mnist["target"]

some_digit = X[4]
some_digit_image = some_digit.reshape(28, 28)

plt.imshow(some_digit_image, cmap="binary")
plt.axis("off")
plt.show()

print(y[4])
```
