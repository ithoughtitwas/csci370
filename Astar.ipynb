This is the A* search. Note that this version uses a search tree and does not start with the search space, Thus, no open and closed lists are used. This keeps some similarity to the algorithm for the uniform cost search.


```python
import copy

# Tho example search trees
#
#            S
#         6/   \4
#         A      B         
#        3|    6/  \2
#         C    D    E
#        4|   7|  5/ \12 
#         G    G  G   H

searchSpace = [('S', 'A', 6), ('A', 'C', 3), ('D', 'G', 7),
('S', 'B', 4), ('C', 'G', 4), ('B', 'D', 6) , ('B', 'E', 2),
('E', 'G', 5), ('E', 'H', 12)]

heuristics = [('A', 5), ('B', 3), ('C', 3), ('D', 6) , ('E', 5),
('H', 4)]

# This function gets the h cost for a node.
def GetHeuristic(heuristics, node):
   if node == 'G':
      return 0
   else:
      x = [item for item in heuristics 
           if item[0] == node]
      return x[0][1]

# Used to sort a list of tuples on the second element.
def GetKey(item):
   return item[1]

# Collects into a list all children of a node.
def GetChildren(space, node):
   list = [item for item in space if item[0] == node]
   return list
```

```python
doSpace = searchSpace

queue = [(['S'], 0)]
goalNode = 'G'

found = False
while queue != [] and found == False:
   print(queue)
   if queue[0][0][-1] == 'G':
      found = True
   else:
      first = queue[0]  # (path, path cost)
      queue = queue[1:]
      children = GetChildren(doSpace, first[0][-1])
      for k in children:  # k is (node, child, branch cost)
         path = copy.deepcopy(first[0])  # existing path
         path.append(k[1])  # add child to new path
         # A path length of two means that the path starts with the start node and contains
         # only the child just added. In this case, the cost is just the branch cost from 
         # the start to the node just added plus the heuristic of the just added node. Else,
         # the last heuristic cost need to be removed (the subtraction in the else part).
         if len(path) == 2:  # the last node before the one just added is the start node
             cost = k[2] + GetHeuristic(heuristics, k[1]) # new path cost
         else:
             cost = first[1] - GetHeuristic(heuristics, path[-2]) + k[2] + \
                    GetHeuristic(heuristics, k[1]) # new path cost
         queue.append((path, cost))
      queue.sort(key=GetKey)

if found == False:
   print('Goal not found')
else:
   print(queue[0])
```
